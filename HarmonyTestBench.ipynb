```python
#Main Window
root=Tk()
#root.geometry('1000x400')
root.title("Harmony")
root.iconbitmap("icons/play.ico")

#Add button
def open_f():
    try:
        top.filename=filedialog.askopenfilename(initialdir="C:\\Users\\Karan Desai\\Harmony\\Music", title="Select the mp3 file", filetypes=(("mp3 files","*.mp3"),("all files","*.*")))
        global fpath
        fpath=top.filename
    except:
        pass
def insert_song():
    
    try:
        mycursor=mydb.cursor()
        sql="INSERT INTO SONG(TRACK_ID,TITLE,DURATION,GENRE,YEAR_RELEASED,ARTIST_ID,ALBUM_ID,FILEPATH) VALUES(%s,%s,%s,%s,%s,%s,%s,%s)"
        val=(track.get(),title.get(),dur.get(),gen.get(),year.get(),artist_id.get(),album_id.get(),fpath)
        mycursor.execute(sql,val)
        messagebox.showinfo("Success!","Song Added")
    except:
        messagebox.showerror("Error","An Error occured, please check the data entered")
def add_song():
    top=Toplevel()
    top.title("Add Songs")
    top.iconbitmap("icons/play.ico")
    
    track=Entry(top,width=50)
    title=Entry(top,width=50)
    dur=Entry(top,width=50)
    gen=Entry(top,width=50)
    year=Entry(top,width=50)
    artist_id=Entry(top,width=50)
    album_id=Entry(top,width=50)
    filepath=Button(top, text='Choose the mp3 file', command=open_f)
    
    track.insert(0,'Track_ID')
    title.insert(0,'Title')
    dur.insert(0,'Duration')
    gen.insert(0,'Genre')
    year.insert(0,'Year')
    artist_id.insert(0,'Artist_ID')
    album_id.insert(0,'Album_ID')
    
    track.grid(row=0,column=0,columnspan=2)
    title.grid(row=1,column=0,columnspan=2)
    dur.grid(row=2,column=0,columnspan=2)
    gen.grid(row=3,column=0,columnspan=2)
    year.grid(row=4,column=0,columnspan=2)
    artist_id.grid(row=5,column=0,columnspan=2)
    album_id.grid(row=6,column=0,columnspan=2)
    filepath.grid(row=7,column=0,columnspan=2)
    
    
    addButton=Button(top,text='Add',command=insert_song).grid(row=8,column=0)
    cancelButton=Button(top,text='Cancel',command=top.destroy).grid(row=8,column=1)
    
    

add=Button(root,text='Add',command=add_song)
add.grid(row=0,column=0)

#Update button
def update_song():
    top=Toplevel()
    top.title("Update Songs")
    top.iconbitmap("icons/play.ico")
    

update=Button(root,text='Update',command=update_song)
update.grid(row=0,column=1)


#Delete button
def delete_song():
    top=Toplevel()
    top.title("Delete Songs")
    top.iconbitmap("icons/play.ico")
    

delete=Button(root,text='Delete',command=delete_song)
delete.grid(row=0,column=2)

#Playlists button
def playlists_dis():
    top=Toplevel()
    top.title("Playlists")
    top.iconbitmap("icons/play.ico")
    

playlist=Button(root,text='Playlists',command=playlists_dis)
playlist.grid(row=0,column=3)

global flag
flag=True
#Frame of Songs
def play_song(fp):
    mixer.init()
    mixer.music.load("Music/"+fp)
    mixer.music.set_volume(0.7)
    if flag==True:
        mixer.music.play()
    elif flag==False:
        mixer.music.unpause()


def pause_song():
    mixer.music.pause()
    
    
sframe=LabelFrame(root,text="Songs".ljust(10), padx=10, pady=10)
sframe.grid(row=1,column=0, columnspan=4, padx=5)
#top_label=Label(sframe,text="T_ID".ljust(50)+"TITLE".ljust(50)+"ARTIST".ljust(50)+"DURATION".ljust(50)+"ALBUM".ljust(50)+"Play/Pause").grid(row=0,column=0,columnspan=4)

mycursor = mydb.cursor()
mycursor.execute("SELECT TRACK_ID, TITLE, A_NAME, DURATION, AL_NAME FROM SONG,ARTIST,ALBUM WHERE SONG.ARTIST_ID=ARTIST.ARTIST_ID AND SONG.ALBUM_ID=ALBUM.ALBUM_ID")
myres=mycursor.fetchall()
ls=myres
label_list=[]
t_id_list=[]
title_list=[]
a_name_list=[]
duration_list=[]
al_name_list=[]
count=0
m=1
t_id_list.append(Label(sframe,text='T_ID',width=3))
title_list.append(Label(sframe,text='TITLE',width=30))
a_name_list.append(Label(sframe,text='ARTIST',width=30))
duration_list.append(Label(sframe,text='DURATION',width=30))
al_name_list.append(Label(sframe,text='ALBUM',width=30))
for i in ls:
    (t_id,title,a_name,duration,al_name)=i
    print(t_id,title,a_name,duration,al_name)
    count+=1
    #label_list.append(Label(sframe,text=f'{t_id}'.ljust(30)+f'{title}'.ljust(30)+f'{a_name}'.ljust(30)+f'{duration}'.ljust(30)+f'{al_name}'.ljust(30),width=300))
    t_id_list.append(Label(sframe,text=f'{t_id}',width=3))
    title_list.append(Label(sframe,text=f'{title}',width=30))
    a_name_list.append(Label(sframe,text=f'{a_name}',width=30))
    duration_list.append(Label(sframe,text=f'{duration}',width=30))
    al_name_list.append(Label(sframe,text=f'{al_name}',width=30))
#for l in label_list:
#    l.grid(row=m,column=0)
#    m+=1

for l in t_id_list:
    l.grid(row=m,column=0)
    m+=1
m=1
for l in title_list:
    l.grid(row=m,column=1)
    m+=1
m=1
for l in a_name_list:
    l.grid(row=m,column=2)
    m+=1
m=1
for l in duration_list:
    l.grid(row=m,column=3)
    m+=1
m=1
for l in al_name_list:
    l.grid(row=m,column=4)
    m+=1


mycursor=mydb.cursor()
mycursor.execute("SELECT FILEPATH FROM SONG")
res2=mycursor.fetchall()
bs=res2
play_button_list=[]
pause_button_list=[]
n=1
o=1
play_button_list.append(Label(sframe,text="Play"))
pause_button_list.append(Label(sframe,text="Pause"))
for j in bs:
    (fp)=j[0]
    print(fp)
    play_button_list.append(Button(sframe, text="▶", command=lambda: play_song(fp)))
    pause_button_list.append(Button(sframe,text="\u23F8", command=pause_song))
for b1 in play_button_list:
    b1.grid(row=n,column=5,padx=5,pady=5)
    n+=1
for b2 in pause_button_list:
    b2.grid(row=o,column=6, padx=5,pady=5)
    o+=1
    
    
    
    
    
'''label_1=Label(sframe,text="Mirrors - Justin Timberlake")
label_1.grid(row=0,column=0)
button_1=Button(sframe, text="▶")
button_2=Button(sframe,text="\u23F8")
button_1.grid(row=0,column=1)
button_2.grid(row=0,column=2)'''

root.mainloop()
```
